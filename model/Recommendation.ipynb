```python
import pandas as pd
df=pd.read_csv("booksgenre.csv",index_col=False)
df=df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
genre=list(df["genres"])
```

```python
genres=[]
for i in genre:
    i=i.split(":")[0][2:-1]
    if "," in i:
        i=i.split(",")
    else:
        genres.append(i)
        continue
    ii=""
    for j in i:
        ii+=j
    genres.append(ii)
```

```python
genres[0]
```

```python
df5=pd.DataFrame()
df5=df4
```

```python
df5["genres"]=genres
```

```python
df5.head()
```

```python
df5['book_author'] = df5['book_author'].map(lambda x: x.split(',')[:3])

```

```python
df5.head()
```

```python
# putting the genres in a list of words
df5['genres'] = df5['genres'].map(lambda x: x.lower().split(' '))

```

```python
df5.head()
```

```python
for index, row in df5.iterrows():
    row['book_author'] = [x.lower().replace(' ','') for x in row['book_author']]
```

```python
df5['Key_words'] = "null"

```

```python
df5.head()
```

## Keyword Extraction from Description

```python
from rake_nltk import Rake
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
words=[]
for index, row in df5.iterrows():
    plot = row['book_description']
    r = Rake()
    r.extract_keywords_from_text(plot)
    key_words_dict_scores = r.get_word_degrees()
    words.append(list(key_words_dict_scores.keys()))

# dropping the Plot column
#df5.drop(columns = ['book_description'], inplace = True)
```

```python
len(words)
```

```python
df5.iloc[1].Key_words
```

```python
df5['Key_words']=words
```

```python
df6=df5[["book_title","book_author","genres","Key_words"]]
```

```python
df6.set_index('book_title', inplace = True)
df6.head()
```

```python
df6.iloc[1].Key_words
```

```python
df6['bag_of_words'] = ''
columns = df6.columns
bag=[]
for index, row in df6.iterrows():
    words = ''
    for col in columns:
        words = words + ' '.join(row[col])+ ' '
    row['bag_of_words'] = words
    bag.append(words)
    
```

```python
df6.head()
```

```python
df6.drop(columns = [col for col in df6.columns if col!= 'bag_of_words'], inplace = True)
```

```python
import pandas as pd
df6=pd.read_csv("reco_df.csv",index_col="book_title")
```

```python
df6.head()
```

```python

count = CountVectorizer()
count_matrix = count.fit_transform(df6['bag_of_words'])


indices = pd.Series(df6.index)
indices[:5]
```

## Cosine Similarity

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
import pandas as pd
def recommendations(title, cosine_sim = cosine_sim):
    
    recommended_books = []
    
    idx = indices[indices == title].index[0]

    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    top_10_indexes = list(score_series.iloc[1:11].index)
    
    for i in top_10_indexes:
        recommended_books.append(list(df6.index)[i])
        
    return recommended_books
```

# Recommendations

```python
recommendations("Harry Potter and the Philosopher's Stone (Harry Potter, #1)")
```

```python
recommendations("Murder on the Orient Express (Hercule Poirot, #10)")
```
